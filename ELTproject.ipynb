# Limpieza y preprocesamiento de datos

```python
df_characteristics=pd.read_csv("games_caracteristicas.csv")
df_characteristics=df_characteristics.drop_duplicates()
```

```python
df_characteristics=df_characteristics.drop(['title', 'thumbnail', 'short_description', 'game_url', 'genre',
       'platform', 'publisher', 'developer', 'release_date',
       'freetogame_profile_url'],axis=1)
```

```python
df_characteristics.rename(columns={"minimum_system_requirements.os":"os","minimum_system_requirements.processor":"processor",
                                "minimum_system_requirements.memory":"memory","minimum_system_requirements.graphics":"graphics",
                                "minimum_system_requirements.storage":"storage"},inplace=True)
```

```python
def c_fecha(valor):
    if "-00" in valor:
        valor=valor.replace("-00","-01")
    return valor
df_games["release_date"]=df_games["release_date"].map(c_fecha)
```

```python
df_games["release_date"]=pd.to_datetime(df_games["release_date"])
df_games.dtypes
```

# subir Dataframes a Postgresql

```python
try:
    chain_connection="postgresql+psycopg2://postgres:root@localhost/games_project"
    engine_connection=create_engine(chain_connection)
    connection=engine_connection.connect()
    print("conexión exitosa")
except Exception as e:
    print(e)
```

```python
table1="games"

try:
    df_sql=df_games.to_sql(table1,connection,if_exists="replace",index=False)
    print("operación exitosa")
except ValueError as v:
    print(v)
except Exception as e:
    print(e)
finally:
    connection.close()
```

```python
table2="games_exinfo"

try:
    df_sql=df_characteristics.to_sql(table2,connection,if_exists="replace",index=False)
    print("operación exitosa")
except ValueError as v:
    print(v)
except Exception as e:
    print(e)
finally:
    connection.close()
```
